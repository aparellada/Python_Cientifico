Un plot simple...

```python
import pylab as pl
pl.plot(range(10))
```

Simple pero con rótulos...

```python
pl.plot(range(10))
pl.xlabel('medido')
pl.ylabel('calculado')
pl.title('Medido vs. Calculado')
pl.grid(True)
```

Varias lineas en un mismo plot...

```python
x = arange(100)
linear = arange(100)
square = [v * v for v in arange(0, 10, 0.1)]
```

```python
lineas = pl.plot(x, linear, x, square)
leyenda = pl.legend(('linear', 'square'))
```

Ajustamos a gusto y *piacere*...

```python
lineas = plot(x, linear, 'g+', x, square, 'ro')
leyenda = legend(('linear', 'square'), loc='upper left')
```

Figuras, plots y subplots...

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
```
